```python
def countplot(dataframe, x_val, plot_title='', figsize=(15,8)):
    plt.figure(figsize=figsize)
    plt.title(plot_title)
    sns.countplot(data=df, x=x_val)
    plt.show()
```

```python
def barplot(dataframe, x_val, y_val):
    sns.barplot(data=dataframe, x=x_val, y=y_val)
    plt.title(x_val.title() + ' vs ' + y_val.title())
    plt.show()
```

Multivariate

```python
fig = plt.figure(figsize = [30, 18]) # defining entire figure size

for i in range(1, len(list(df.columns))):
    plt.subplot(8, 3, i) # makes subplot
    
    fig.tight_layout(pad=2) # gives space between the subplots
    df_new = df.iloc[:,i] # making new dataframe using slicing technique to remote the unwanted columns
    
    plt.suptitle('Comparision of all Features') # main title
    plt.title(list(df.columns)[i]) # title for each subplots
    
    sns.boxenplot(df_new) # making boxen plot

plt.show() # displaying the plot
```
